# Feature Engineering

```python
train_set.isnull().sum()
```

```python
def question_feature_getter(df):
    columns = []
    for col in df.columns:
        if col[:4] == 'SORU':
            columns.append(col)
    return columns
```

```python
train_question_cols = question_feature_getter(train_set)
```

```python
train_question_df = train_set.loc[:, train_question_cols]
```

```python
train_set.drop(train_question_cols, axis=1, inplace=True)
```

```python
test_question_cols = question_feature_getter(test_set)
```

```python
test_question_df = test_set.loc[:, test_question_cols]
```

```python
test_set.drop(test_question_cols, axis=1, inplace=True)
```

```python
train_question_df['q_non_null_count'] = train_question_df.apply(lambda row: row.notnull().sum(), axis=1)
test_question_df['q_non_null_count'] = test_question_df.apply(lambda row: row.notnull().sum(), axis=1)
```

```python
train_set = train_set.assign(CEVAPLANAN_ANKET_SAYISI = train_question_df['q_non_null_count'])
test_set = test_set.assign(CEVAPLANAN_ANKET_SAYISI = test_question_df['q_non_null_count'])
```

```python
drop_son = train_set.filter(like='SON').columns
drop_tutar = train_set.filter(like='TUTAR').columns
train_set = train_set.drop(drop_son, axis=1)
train_set = train_set.drop(drop_tutar, axis=1)
train_set = train_set.drop(['AKTIF_ILK_POLICE_RG', 'ANAPARA', 'GETIRI'], axis=1)

drop_son = test_set.filter(like='SON').columns
drop_tutar = test_set.filter(like='TUTAR').columns
test_set = test_set.drop(drop_son, axis=1)
test_set = test_set.drop(drop_tutar, axis=1)
test_set = test_set.drop(['AKTIF_ILK_POLICE_RG', 'ANAPARA', 'GETIRI'], axis=1)
```

```python
train_set.columns
```

```python
train_set.drop('MUSTERI_ID', axis=1, inplace=True)
test_set.drop('MUSTERI_ID', axis=1, inplace=True)
```

```python
train_set.columns
```

```python
y_train = train_set['LABEL']
X_train = train_set.drop(['LABEL'], axis=1)
```

# Encoding catogerical features

```python
X_train.info()
```

Splitting catogeric and numeric features

```python
# Defining catogeric and numeric features
catogeric_features = ['FLAG', 'PP_CINSIYET', 'PP_MESLEK', 'PP_MUSTERI_SEGMENTI',
                      'PP_UYRUK', 'IL']
numeric_features = [col for col in X_train.columns if col not in catogeric_features]
```

```python
# Splitting catogeric and numeric features on training set
catogeric_X_train = X_train[catogeric_features]
numeric_X_train = X_train.drop(catogeric_features, axis=1)
```

```python
# Splitting catogeric and numeric features on test set
catogeric_test = test_set[catogeric_features]
numeric_test = test_set.drop(catogeric_features, axis=1)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one_hot_encoder = OneHotEncoder(drop="first", handle_unknown="ignore", sparse_output=False)
```

```python
catogeric_X_train_encoded_nd = one_hot_encoder.fit_transform(catogeric_X_train)
catogeric_X_train_encoded = pd.DataFrame(catogeric_X_train_encoded_nd, columns=one_hot_encoder.get_feature_names_out(catogeric_features))
```

```python
catogeric_test_encoded_nd = one_hot_encoder.transform(catogeric_test)
catogeric_test_encoded = pd.DataFrame(catogeric_test_encoded_nd, columns=one_hot_encoder.get_feature_names_out(catogeric_features))
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
y_train_encoded = label_encoder.fit_transform(y_train)
```

Concatenating catogeric and numeric features

```python
X_train_encoded = pd.concat([catogeric_X_train_encoded, numeric_X_train], axis=1)
test_set_encoded = pd.concat([catogeric_test_encoded, numeric_test], axis=1)
```
